```python
import os
os.listdir('/kaggle/input/titanic-task5')

import pandas as pd

df = pd.read_csv('/kaggle/input/titanic-task5/train.csv')
df.head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10,6)
sns.set(style='whitegrid')


```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Pclass vs Survived")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Sex vs Survived")
plt.show()
```

```python
df['Age'].hist(bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Females survived more compared to males.

First class passengers had better survival chances.

Age has missing values  might require imputation.

Fare is positively correlated with survival.
```
